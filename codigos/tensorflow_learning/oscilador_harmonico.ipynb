```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 1
t_final = 10*2*np.pi/w
pontos = 1000

h = t_final/pontos
```

```python
h
```

```python
tt = np.linspace(0,t_final,pontos)
xx = np.empty(len(tt))
vv = np.empty(len(tt))
xx[0] = 5
vv[0] = -2
```

```python
for i in range(len(xx)-1):
    vv[i+1] = vv[i] - w**2*xx[i]*h
    xx[i+1] = xx[i] + vv[i]*h
```

```python
plt.plot(tt,xx)
plt.xlabel('tempo (s)')
plt.ylabel('posio (m)')
plt.title('Oscilao de um oscilador harmnico')
plt.grid()
plt.savefig('imagens/Oscilao de um oscilador harmonico euler')
```

```python
plt.plot(xx,vv)
plt.xlabel('posio (m)')
plt.ylabel('momento (kg $\cdot$ m/s)')
plt.grid()
plt.title('Diagrama de fase')
plt.savefig('imagens/diagrama de fase euler')
```
