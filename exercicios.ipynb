```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn

import pandas as pd  
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
def R2(y_data, y_model):
    return 1 - np.sum((y_data - y_model) ** 2) / np.sum((y_data - np.mean(y_data)) ** 2)
def MSE(y_data,y_model):
    n = np.size(y_model)
    return np.sum((y_data-y_model)**2)/n
```

```python
x = np.random.rand(100,1)
y = 2.+5*x**2+.1*np.random.randn(100,1)
```

1. Write your own code (following the examples under the regression notes) for computing the parametrization of the data set fitting a second-order polynomial.

```python
p = 3 # b0,b1,b2
n = len(y)
X = np.zeros((n,p))

for i in range(n):
    for j in range(p):
        X[i][j] = x[i]**j
X
```

```python
# separar entre dados de teste e dados de treino

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
```

embaixo tá errado porque o $\beta$ é dado por $$\beta = (X^T X)^{-1} X^T y$$

```python
beta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
beta
```

```python
ytilde = X_train @ beta
```

```python
print("Training R2")
print(R2(y_train,ytilde))
print("Training MSE")
print(MSE(y_train,ytilde))
ypredict = X_test @ beta
print("Test R2")
print(R2(y_test,ypredict))
print("Test MSE")
print(MSE(y_test,ypredict))
```
